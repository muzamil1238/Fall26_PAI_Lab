# Dynamic N-Queens Problem Solver

## Features:
- **Dynamic board size**: Input any value of n

## How it works:
The N-Queens problem is to place n queens on an n√ón chessboard so that no two queens threaten each other. This means:
- No two queens share the same row
- No two queens share the same column
- No two queens share the same diagonal

```python
def solve_n_queens():
    n = int(input("Enter value of n: "))
    board = [-1] * n
    solutions = []

    def is_safe(row, col):
        for i in range(row):
            if board[i] == col or abs(board[i] - col) == abs(i - row):
                return False
        return True

    def print_board():
        print()
        print("  " + "---" * n)
        for r in range(n):
            row_line = "|"
            for c in range(n):
                if board[r] == c:
                    row_line += " Q |"
                else:
                    row_line += " . |"
            print(row_line)
            print("  " + "---" * n)
        print()

    def dfs(row):
        if row == n:
            solutions.append(board[:])
            print("Solution Found:")
            print_board()
            return

        for col in range(n):
            if is_safe(row, col):
                board[row] = col
                dfs(row + 1)
                board[row] = -1

    dfs(0)

    if not solutions:
        print("No solution found")
    else:
        print("Total Solutions:", len(solutions))


solve_n_queens()
```
